```python
# Necessary imports
import psycopg2
from psycopg2.extras import execute_values
import pandas as pd
import spacy
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity, linear_kernel
from string import punctuation
from scipy.sparse import csr_matrix
import unicodedata
import re
import numpy as np
import os
import sys
```

```python
# Load the Portuguese language model
nlp = spacy.load('pt_core_news_lg', exclude=["tok2vec", "tagger", "parser", "ner", "attribute_ruler"])

# Define stop words and unwanted words
STOP_WORDS = set(nlp.Defaults.stop_words)
unwanted_words = ['processo', 'cliente', 'estratégia', 'Avaliar', 'Responsabilidades', 'gestor', 'MRLD', 'risco', 'específico', 'tomador', 'Risco']
STOP_WORDS -= set(unwanted_words)
```

```python
def remove_accents(input_str):
    nfkd_form = unicodedata.normalize('NFKD', input_str)
    return "".join([c for c in nfkd_form if not unicodedata.combining(c)])
```

```python
def preprocess_text(text):
    if isinstance(text, str):
        # Remover acentos
        text = remove_accents(text)
        # Remover tags HTML
        text = re.sub('<.*?>', '', text)
        doc = nlp(text)
        doc = [
            token.lemma_.lower()
            for token in doc
            if token.text.lower() not in STOP_WORDS
            and token.text not in unwanted_words
            and token.text not in punctuation
            and not token.is_punct
            and not token.is_space
            and token.lemma_ != '-PRON-'
        ]
        preprocessed_text = ' '.join(doc)
        return preprocessed_text
    else:
        return ''
```

```python
# Read the data
query = """
    SELECT distinct
        CONCAT(cd_prf_depe, cd_atvd_plnj, cd_qsto, cd_qst, cd_atnt_qst) as id,
        ano,
        cd_papl_trb, nm_papl_trb,
        cd_prf_depe,
        cd_atvd_plnj,
        cd_qsto, tx_qsto,
        cd_qst,
        cd_atnt_qst,
        tx_cmpr_rpst_atnt
    FROM mrld_e_perdas.compara_resposta 
    WHERE (cd_papl_trb = 1543 and tx_cmpr_rpst_atnt is not null) or (cd_papl_trb <> 1543);
"""
df = pd.read_sql(query, engine)
```

```python
# Garantir que os IDs sejam strings e remover espaços em branco
df['id'] = df['id'].astype(str).str.strip()

# Pré-processar os textos
df['tx_cmpr_rpst_atnt'] = df['tx_cmpr_rpst_atnt'].fillna('').astype(str)
df['tx_cmpr_rpst_atnt_processado'] = df['tx_cmpr_rpst_atnt'].apply(preprocess_text)

# Vetorização TF-IDF usando matriz esparsa
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(df['tx_cmpr_rpst_atnt_processado'])

# Função para computar similaridades acima de um certo limiar
def get_similarities_above_threshold(tfidf_matrix, threshold=0.1):
    # Usando kernel linear para produto interno
    cosine_similarities = linear_kernel(tfidf_matrix, tfidf_matrix)

    # Aplicar máscara para valores acima do limiar
    similarity_matrix = np.where(cosine_similarities > threshold, cosine_similarities, 0)

    # Converter para matriz esparsa
    sparse_sim_matrix = csr_matrix(similarity_matrix)

    return sparse_sim_matrix

# Obter a matriz de similaridade esparsa
sparse_sim_matrix = get_similarities_above_threshold(tfidf_matrix, threshold=0.9)

# Converter a matriz esparsa em um DataFrame
sim_indices = sparse_sim_matrix.nonzero()
sim_scores = sparse_sim_matrix.data

similarity_df = pd.DataFrame({
    'id_idx': sim_indices[0],
    'id_compare_idx': sim_indices[1],
    'similarity': sim_scores
})

# Mapear os índices para os IDs reais
similarity_df['id'] = df['id'].values[similarity_df['id_idx']]
similarity_df['id_compare'] = df['id'].values[similarity_df['id_compare_idx']]

# Remover colunas de índice
similarity_df = similarity_df.drop(columns=['id_idx', 'id_compare_idx'])

# Garantir que os IDs sejam comparados corretamente
similarity_df['id'] = similarity_df['id'].astype(str).str.strip()
similarity_df['id_compare'] = similarity_df['id_compare'].astype(str).str.strip()

# Remover auto-similaridades
similarity_df = similarity_df[similarity_df['id'] != similarity_df['id_compare']]

# Remover duplicatas onde id e id_compare são invertidos
similarity_df['id_pair'] = similarity_df.apply(
    lambda row: '_'.join(sorted([row['id'], row['id_compare']])), axis=1
)
similarity_df = similarity_df.drop_duplicates(subset='id_pair')
similarity_df = similarity_df.drop(columns=['id_pair'])

# **Mesclar para obter as colunas adicionais para 'id'**
similarity_df = similarity_df.merge(
    df[['id', 'ano', 'cd_papl_trb', 'nm_papl_trb', 'cd_prf_depe', 'cd_atvd_plnj', 'cd_qsto',  'tx_qsto', 'cd_qst', 'cd_atnt_qst', 'tx_cmpr_rpst_atnt']],
    on='id', how='left'
)
similarity_df = similarity_df.rename(
    columns={
        'ano' :'ano_id',
        'cd_papl_trb': 'cd_papl_trb_id',
        'nm_papl_trb': 'nm_papl_trb_id',
        'tx_qsto': 'tx_qsto_id',
        'cd_prf_depe': 'cd_prf_depe_id',
        'cd_atvd_plnj': 'cd_atvd_plnj_id',
        'cd_qsto': 'cd_qsto_id',
        'cd_qst': 'cd_qst_id',
        'cd_atnt_qst': 'cd_atnt_qst_id',
        'tx_cmpr_rpst_atnt': 'tx_cmpr_rpst_atnt_original'
    }
)

# **Mesclar para obter as colunas adicionais para 'id_compare'**
df_compare = df[['id', 'ano', 'cd_papl_trb', 'nm_papl_trb', 'cd_prf_depe', 'cd_atvd_plnj', 'cd_qsto', 'tx_qsto', 'cd_qst', 'cd_atnt_qst', 'tx_cmpr_rpst_atnt']].rename(
    columns={
        'id': 'id_compare',
        'ano' :'ano_compare',
        'cd_papl_trb': 'cd_papl_trb_compare',
        'nm_papl_trb': 'nm_papl_trb_compare',
        'tx_qsto': 'tx_qsto_compare',
        'cd_prf_depe': 'cd_prf_depe_compare',
        'cd_atvd_plnj': 'cd_atvd_plnj_compare',
        'cd_qsto': 'cd_qsto_compare',
        'cd_qst': 'cd_qst_compare',
        'cd_atnt_qst': 'cd_atnt_qst_compare',
        'tx_cmpr_rpst_atnt': 'tx_cmpr_rpst_atnt_compare'
    }
)
similarity_df = similarity_df.merge(
    df_compare, on='id_compare', how='left'
)

# Definir observações com base nos limiares de similaridade
def define_observation(similarity):
    if similarity > 0.9:
        return 'Probabilidade de duplicação de tema'
    elif 0.7 < similarity <= 0.9:
        return 'Existe tema semelhante ao cadastrado'
    elif 0.5 < similarity <= 0.7:
        return 'Similaridade entre 50% e 70%, alguns textos em comum'
    elif 0.2 < similarity <= 0.5:
        return 'Similaridade entre 20% e 50%, poucas palavras em comum'
    else:
        return 'Similaridade abaixo de 20%, geralmente a pesquisa possui poucas palavras para análise da contextualização'

similarity_df['Obs'] = similarity_df['similarity'].apply(define_observation)

# Ordenar e exibir os resultados
similarity_df = similarity_df.sort_values(by='similarity', ascending=False)

# Selecionar as colunas desejadas para exibição
cols_to_display = [
    'cd_papl_trb_id', 'nm_papl_trb_id', 'id', 'id_compare', 'similarity', 'Obs',
    'cd_prf_depe_id', 'cd_atvd_plnj_id', 'cd_qsto_id', 'tx_qsto_id', 'cd_qst_id', 'cd_atnt_qst_id',
    'cd_papl_trb_compare', 'nm_papl_trb_compare', 'cd_prf_depe_compare', 'cd_atvd_plnj_compare', 'cd_qsto_compare', 'tx_qsto_compare', 
    'cd_qst_compare', 'cd_atnt_qst_compare', 'tx_cmpr_rpst_atnt_original', 'tx_cmpr_rpst_atnt_compare'
]
```

```python
# Fazer rollback para garantir que a conexão esteja em um estado limpo 
conn.rollback() 
# Agora faça a verificação e adição da coluna fora de uma transação ativa 
with conn.cursor() as cur:
    # Verificar se a coluna já existe 
    cur.execute(""" 
    SELECT column_name FROM information_schema.columns 
    WHERE table_name = 'compara_resposta' AND column_name = 'tot_similaridades'; 
    """) 
   
    result = cur.fetchone()
    # Se a coluna não existir, adicioná-la 
    if not result: 
        cur.execute(""" 
        ALTER TABLE mrld_e_perdas.compara_resposta ADD COLUMN tot_similaridades INT; 
        """) 
        conn.commit()
```

```python
data = [
    (row['tot_similaridades'], row['cd_prf_depe'], row['cd_atvd_plnj'], row['cd_qsto'], row['cd_qst'], row['cd_atnt_qst'])
    for index, row in df.iterrows()
]
```

```python
# Gerar o comando de atualização em lote
query_update = """
    UPDATE mrld_e_perdas.compara_resposta as t
    SET tot_similaridades = v.tot_similaridades
    FROM (VALUES %s) as v(tot_similaridades, cd_prf_depe, cd_atvd_plnj, cd_qsto, cd_qst, cd_atnt_qst)
    WHERE t.cd_prf_depe = v.cd_prf_depe
      AND t.cd_atvd_plnj = v.cd_atvd_plnj
      AND t.cd_qsto = v.cd_qsto
      AND t.cd_qst = v.cd_qst
      AND t.cd_atnt_qst = v.cd_atnt_qst and v.tot_similaridades > 0;
"""
```

```python
# Gerar o comando de atualização em lote
query_update2 = """
    UPDATE mrld_e_perdas.compara_resposta as t
    SET tot_similaridades = 0
    FROM (VALUES %s) as v(tot_similaridades, cd_prf_depe, cd_atvd_plnj, cd_qsto, cd_qst, cd_atnt_qst)
    ;
"""
```

```python
# Executar o update em lote
with conn.cursor() as cur:
    execute_values(cur, query_update, data)
    conn.commit()
```
