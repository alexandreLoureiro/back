````python
def extract_text_from_pdf(pdf_path):
    try:
        return extract_text(pdf_path)
    except Exception as e:
        # Logging do erro se necessário
        return ""


# Função para limpar o texto extraído
def clean_text(text):
    if not isinstance(text, str):
        return ""
    texto_limpo = re.sub(r'\n\s*\n+', '\n\n', text)  # Remove múltiplas quebras
    texto_limpo = re.sub(r'[ ]{2,}', ' ', texto_limpo)  # Remove espaços múltiplos
    linhas = [linha.strip() for linha in texto_limpo.split('\n')]
    linhas = [linha for linha in linhas if linha]  # linhas não vazias
    return '\n'.join(linhas)


df_arqs['extracted_text'] = df_arqs['NOME_ARQUIVO'].apply(
    lambda x: clean_text(extract_text_from_pdf(os.path.join(pdf_path_base, x))) if isinstance(x, str) else ''
)

````

````python

def extract_text_from_pdf(pdf_path):
    return extract_text(pdf_path)

# Função para limpar o texto extraído
def clean_text(text):
    # Remover múltiplas quebras de linha consecutivas
    texto_limpo = re.sub(r'\n\s*\n+', '\n\n', text)
    # Remover múltiplos espaços dentro das linhas
    texto_limpo = re.sub(r'[ ]{2,}', ' ', texto_limpo)
    # Remover espaços em branco no início e no fim das linhas
    linhas = texto_limpo.split('\n')
    linhas = [linha.strip() for linha in linhas]
    # Remover linhas completamente vazias
    linhas = [linha for linha in linhas if linha != '']
    # Reunir as linhas em um único texto
    return '\n'.join(linhas)

# Adicionar a coluna 'extracted_text' ao DataFrame, tratando valores nulos na coluna 'NOME_ARQUIVO'
df_csv['extracted_text'] = df_csv['NOME_ARQUIVO'].apply(lambda x: clean_text(extract_text_from_pdf(os.path.join(pdf_path_base, x))) if pd.notnull(x) and isinstance(x, str) else '')
````

````python
def assistenteKmpg(prompt):
    
    url = 'http://acs-assist-vinculo-plan.nia.desenv.bb.com.br/acs/llms/agent'
    resp = requests.post(url, data=json.dumps({
        "data": {
            "input": prompt,
            "intents": [{}],
            "entities": [{}],
            "context": {
                "config": {
                    "top_p": 0.95,
                    "max_tokens": 4000,
                    "temperature": 0.2
                }
            }
        }
    }), headers={'Content-type': 'application/json', 'accept': 'application/json'})

    # Convertendo a resposta em um dicionário
    resp_dict = json.loads(resp.text)

    # Extraindo o status
    status = resp_dict.get('status')
    print('status, retorno 200 ok, diferente contactar nia ', status)
    # simular retorno no swagger : http://acs-assist-vinculo-plan.nia.hm.bb.com.br/nia_semantica_acs/docs

     # Extraindo o resultado da resposta
    texto_resposta = resp_dict["data"]["context"]["prompt_response"]["acs_llm_prompt_execution"]["result"]
    
    # Removendo marcadores de código e espaços em branco
    texto_resposta = texto_resposta.strip()
    if texto_resposta.startswith('```json'):
        texto_resposta = texto_resposta[len('```json'):].strip()
    if texto_resposta.endswith('```'):
        texto_resposta = texto_resposta[:-len('```')].strip()

    # Tentando carregar o JSON
    try:
        resposta_json = json.loads(texto_resposta)
        
    except json.JSONDecodeError as e:
        print(f"Erro ao decodificar o JSON: {e}")
        # Como alternativa, usar a expressão regular se a decodificação falhar
        match = re.search(r'\{.*\}', texto_resposta, re.DOTALL)
        if match:
            json_part = match.group(0)
            try:
                resposta_json = json.loads(json_part)
            except json.JSONDecodeError as e:
                print(f"Erro ao decodificar o JSON após regex: {e}")
                resposta_json = {}
        else:
            resposta_json = {}
            
    # Criando o DataFrame
    print('resposta_json ', resposta_json)
    df = pd.DataFrame([resposta_json])
    
    # Removendo a coluna 'informacao' se existir
    if 'informacao' in df.columns:
        df = df.drop(columns=['informacao'])
    return df
````

````python
resultados2 = []
for index, row in df_csv.iterrows():
    id_tabela = row['id']
    nome_relatorio = row['RELATORIO']
    relatorio = row['extracted_text']
    conf_constatacoes = row['CONSTATACAO']
    conf_recomendacao = row['RECOMENDACAO']
    conf_implementacao = str(row['PREVISAO_IMPLEMENTACAO'])
    
    prompt = f""" Você receberá um relatório de auditoria externa: {relatorio}

    quero que voce analise se nesse relatorio tem o registro dessa constatacao: {conf_constatacoes} e esta recomendação {conf_recomendacao}    
    caso encontre a constatacao ou a recomendação informe a página ou de uma dica de onde está para facilitar a visualizção do funcionário     
    informar se a data de implementação na constatacao é {conf_implementacao}    
    
    2. **Retornar o resultado no seguinte formato JSON:**
    
          {{
                "achou_constatacao": "Sim/Não",
                "achou_recomendacao": "Sim/Não",
                "dica_localizar_constatacao": "caso encontre a constatacao ou a recomendação informe a página ou de uma dica de onde está para facilitar a visualizção do funcionário",
                "dica_localizar_recomendacao": "caso encontre a constatacao ou a recomendação informe a página ou de uma dica de onde está para facilitar a visualizção do funcionário",
                "conf_implementacao": "Correta, sim ou não",
                "racional": "Racional por que considera que o achado foi regularizado durante o trabalho, incluindo a declaração explícita do auditor ou explicação do porquê não foi encontrado",
                "resumo_constatacao": "Resumo da constatacao",
                "resumo_recomendacao": "Resumo da recomendacao"
            }}
        OBS.: Envie o JSON em formato puro sem vestígios de markdown, remove a marcação abaixo:
        [REMOVER]:
        ```json
       ```
    """
    resumo_rel_proc_list = assistenteKmpg(prompt)
    resumo_rel_proc_list['id'] = id_tabela
    resumo_rel_proc_list['nome_relatorio'] = nome_relatorio
    resumo_rel_proc_list['relatorio'] = relatorio
    resumo_rel_proc_list['conf_constatacoes'] = conf_constatacoes
    resumo_rel_proc_list['conf_recomendacao'] = conf_recomendacao
    resumo_rel_proc_list['conf_implementacao'] = conf_implementacao
    
    resultados2.append(resumo_rel_proc_list)
    
# Concatenar todos os resultados em um DataFrame
df_resultados2 = pd.concat(resultados2, ignore_index=True)
````
